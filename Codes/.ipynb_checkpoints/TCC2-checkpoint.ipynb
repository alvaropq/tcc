```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Base_final.csv')
X = dataset.drop('class',axis=1)
```

```python
dataset
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(X)
```

```python
X_train_minmax
```

```python
clusters = len(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_r = pca.fit(X_train_minmax).transform(X_train_minmax)
```

```python
X_r
```

```python
y = dataset['class']
```

```python
clusters
```

```python
target_names = dataset['class']
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure()
lw = clusters

for i, target_name in zip( dataset['class'], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA')
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X_train_minmax, method  = "ward"), labels = list(dataset['class']))
plt.xticks(rotation=90)
plt.title('Dendrograma')
plt.xlabel('Cepas')
plt.ylabel('Distancia Euclidiana')
plt.savefig('Dendograma.pdf', format='pdf')
plt.show()
```

```python
from scipy.spatial import distance
dist = pd.DataFrame()
i = 0;
j = 0;
while i < clusters:
    linha = []
    a = tuple(X_train_minmax[i])
    while j< clusters:
        b = tuple(X_train_minmax[j])
        dst = distance.euclidean(a, b)
        linha.append(dst)
        j = j+1
    coluna = []
    coluna.append(linha)
    aux = pd.DataFrame(coluna)
    dist= pd.concat([dist,aux])
    i = i+1
    j = 0
```

```python
dist = dist.reset_index(drop=True)
```

```python
mapping = {}
for i, row in dataset.iterrows():
        mapping[i] = row['class']
```

```python
mapping
```

```python
dist.rename(columns=mapping)
```

```python
dist['class'] = dataset['class']
```

```python
import seaborn as sb 
fig, ax = plt.subplots(figsize=(10, 8))
# mask
mask = np.triu(np.ones_like(dist, dtype=np.bool))
# adjust mask and df
mask = mask[1:, :-1]
corr = dist.iloc[1:,:-1].copy()
# plot heatmap
sb.heatmap(corr, mask=mask, annot=True, fmt=".1f", cmap='coolwarm',
           vmin=0, vmax=10, cbar_kws={"shrink": .8})
# yticks
plt.yticks(rotation=0)
plt.savefig('heatmap.pdf', format='pdf')
plt.show()
```
